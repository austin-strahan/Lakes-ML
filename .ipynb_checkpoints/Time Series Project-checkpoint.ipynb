## Arima

```python
arima = clean_df[['date','mc_ppb']].dropna()
arima
```

```python
plt.style.use('fivethirtyeight')

# Plot and show the time series on axis ax1
fig, ax1 = plt.subplots()
arima.plot(ax=ax1, figsize=(14,12))
plt.show()
```

```python
# Split the data into a train and test set
arima_train = arima.loc[:'2013']
arima_test = arima.loc['2004':]

# Create an axis
fig, ax = plt.subplots()

# Plot the train and test sets on the axis ax
arima_train.plot(ax=ax, figsize=(12,10))
arima_test.plot(ax=ax)
plt.title('train - test split of the monthly production of candy in US')
plt.xlabel('Date')
plt.ylabel('Production')
plt.show()
```

## Dickey Fuller Test

```python
from statsmodels.tsa.stattools import adfuller
adf, pval, usedlag, nobs, crit_vals, icbest =  adfuller(arima.mc_ppb.values)
print('ADF test statistic:', adf)
print('ADF p-values:', pval)
print('ADF number of lags used:', usedlag)
print('ADF number of observations:', nobs)
print('ADF critical values:', crit_vals)
print('ADF best information criterion:', icbest)
```

## Since p value is less than 0.05, our data is stationary
